# **Building the API**

```python
import gradio as gr
import json
```

```python
# Now create the function
def get_response(input_data):
  data = json.loads(input_data)
  text = data["query"]

  hii = tokenizer(text, return_tensors="pt").to("cuda")
  output = model.generate(**hii, max_new_tokens=512)
  response = tokenizer.decode(output[0])

  return {"response":response}
```

```python
demo = gr.Interface(fn = get_response, inputs="json", outputs="json")
```

```python
demo.queue().launch(share=True, debug=True, show_error=True)
```
